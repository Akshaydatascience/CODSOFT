```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
work5 = pd.read_csv('creditcard.csv')
```

```python
work5.head()
```

```python
work5.info()
```

```python
work5.isnull().sum()
```

```python
work5['Class'].value_counts()
```

```python
legit = work5[work5.Class == 0]
fraud = work5[work5.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
work5.groupby('Class').mean()
```

```python
legit_sample = legit.sample(n=492)
```

```python
newdataset = pd.concat([legit_sample, fraud], axis=0)
```

```python
newdataset.head()
```

```python
newdataset['Class'].value_counts()
```

```python
newdataset.groupby('Class').mean()
```

```python
x = newdataset.drop(columns='Class', axis=1)
y = newdataset['Class']
```

```python
print(x)
```

```python
print(y)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, stratify=y, random_state=2)
```

```python
print(x.shape, x_train.shape, x_test.shape)
```

```python
model = LogisticRegression()
```

```python
model.fit(x_train, y_train)
```

```python
x_train_prediction = model.predict(x_train)
training_data_accuracy = accuracy_score(x_train_prediction, y_train)
```

```python
print('Accuracy on Training data : ', training_data_accuracy)
```

```python
x_test_prediction = model.predict(x_test)
test_data_accuracy = accuracy_score(x_test_prediction, y_test)
```

```python
print('Accuracy score on Test Data : ', test_data_accuracy)
```
